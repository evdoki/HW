```python
import pandas as pd

url = "https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/new_year_film.csv"
c=pd.read_csv(url)
c
```

```python
p=c.ranking.str.replace(',','.')

round(pd.to_numeric(p.str.replace('\D', '.')).mean(),2)
```
